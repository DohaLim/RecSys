# 협업 필터링 추천 시스템

```python
import pandas as pd
```

### 데이터 준비

```python
df_movies = pd.read_csv('data/movie-1m/movies.dat', sep='::', encoding="ISO-8859-1")
df_ratings = pd.read_csv('data/movie-1m/ratings.dat', sep='::', encoding="ISO-8859-1")
df_users = pd.read_csv('data/movie-1m/users.dat', sep='::', encoding="ISO-8859-1")
```

```python
df_movies.columns = ['MovieID', 'Title', 'Genres']
movie_first = pd.DataFrame({'MovieID': '1', 'Title':'Toy Story (1995)', 'Genres':"Animation|Children's|Comedy"}, index=[0])
df_movies = pd.concat([movie_first, df_movies[:]]).reset_index(drop=True)
```

```python
df_ratings.columns = ['UserID', 'MovieID', 'Rating', 'Timestamp']
rating_first = pd.DataFrame({'UserID':'1', 'MovieID':'1193', 'Rating':'5', 'Timestamp':'978300760'}, index=[0])
df_ratings = pd.concat([rating_first, df_ratings[:]]).reset_index(drop=True)
```

```python
df_users.columns = ['UserID','Gender','Age','Occupation','Zip-code']
user_first = pd.DataFrame({'UserID':'1', 'Gender':'F', 'Age':'1.1', 'Occupation':'10', 'Zip-code':'48067'}, index=[0])
df_users = pd.concat([user_first, df_users[:]]).reset_index(drop=True)
```

```python
# 반드시 merge하기 전에 컬럼의 datatype이 일치하는지 확인!!
df_movies['MovieID'] = df_movies['MovieID'].astype(int)
df_ratings['MovieID'] = df_ratings['MovieID'].astype(int)

df_users['UserID'] = df_users['UserID'].astype(int)
df_ratings['UserID'] = df_ratings['UserID'].astype(int)
```

```python
df_merged = pd.merge(df_ratings, df_movies, on='MovieID', how='left')
df_merged = pd.merge(df_merged, df_users, on='UserID', how='left')
```

```python
df_merged
```

### 데이터 전처리

```python
df_merged = df_merged.sort_values(by=['UserID', 'Timestamp'], ignore_index=True)
```

```python
# Age 오류 처리
df_merged['Age'] = df_merged['Age'].replace('1.1', 1)
```

```python
# Gender 값 변경
df_merged['Gender'] = df_merged['Gender'].replace('F', 4)
df_merged['Gender'] = df_merged['Gender'].replace('M', 3)
```

```python
# Genre pivot
df_genre = df_merged['Genres'].str.get_dummies("|")
df_pivot = pd.concat([df_merged, df_genre], axis=1)
```

```python
# 불필요한 columns 제거
df_pivot = df_pivot.drop(['Title', 'Genres', 'Zip-code'], axis=1)
```
